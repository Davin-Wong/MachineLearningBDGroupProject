```python
# CMPE 188 Project: The Problems of American Health Care
# Team: Asian Analysis
# Kevin Prakasa
# Davin Wong
# Kevin Wong

print(__doc__)

import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt

# Reading the Money Spent on Health Care csv and the Life Expectancy csv
MoneySpent = pd.read_csv('health_spending_per_cap.csv')
LifeExpectancy = pd.read_csv('life_expect_birth.csv')
```

```python
# Printing out the Money Spent Data Set
MoneySpent
```

```python
# Printing out the Life Expectancy Data Set
LifeExpectancy
```

```python
# Removing Unneeded Columns from Both Sets
MoneySpent = MoneySpent[["LOCATION", "TIME", "Value"]]
LifeExpectancy = LifeExpectancy[["LOCATION", "TIME", "Value"]]

# New DataFrame for estimated world avg
EstWorldAvg = pd.merge(MoneySpent, LifeExpectancy, on=['LOCATION', 'TIME'], how='left')
EstWorldAvg = EstWorldAvg.dropna()
```

```python
# Printing out EstWorldAvg DataFrame
EstWorldAvg
```

```python
# Export the EstWorldAvg DataFrame into a csv file
EstWorldAvg.to_csv('EstWorldAvgPreprocessed.csv', index = False)
```

```python
# Create 5 new DataFrames for major countries of our choosing including the USA: USA, CANADA, JAPAN, FRANCE, and the UK
# This will be our tweak
# First, we will compare the countries to each other to rank them from best to worst
# Then, we will compare them to our EstWorldAvg DataFrame
Canada = EstWorldAvg[EstWorldAvg['LOCATION'] == 'CAN']
France = EstWorldAvg[EstWorldAvg['LOCATION'] == 'FRA']
Japan = EstWorldAvg[EstWorldAvg['LOCATION'] == 'JPN']
UK = EstWorldAvg[EstWorldAvg['LOCATION'] == 'GBR']
US = EstWorldAvg[EstWorldAvg['LOCATION'] == 'USA']
```

```python
# Set x and y values for all DataFrames
x_canada = Canada['Value_x']
y_canada = Canada['Value_y']

x_france = France['Value_x']
y_france = France['Value_y']

x_japan = Japan['Value_x']
y_japan = Japan['Value_y']

x_uk = UK['Value_x']
y_uk = UK['Value_y']

x_us = US['Value_x']
y_us = US['Value_y']

x_world = EstWorldAvg['Value_x']
y_world = EstWorldAvg['Value_y']
```

```python
# Fix x and y values
x_canada = x_canada.values.reshape(-1,1)
x_france = x_france.values.reshape(-1,1)
x_japan = x_japan.values.reshape(-1,1)
x_uk = x_uk.values.reshape(-1,1)
x_us = x_us.values.reshape(-1,1)
x_world = x_world.values.reshape(-1,1)

np.isnan(y_canada)
y_canada = y_canada.astype(int)

np.isnan(y_france)
y_france = y_france.astype(int)

np.isnan(y_japan)
y_japan = y_japan.astype(int)

np.isnan(y_uk)
y_uk = y_uk.astype(int)

np.isnan(y_us)
y_us = y_us.astype(int)

np.isnan(y_world)
y_world = y_world.astype(int)
```

```python
# Make Training and Test Sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_world, y_world, test_size=0.3, random_state=0)
X_trainCAN, X_testCAN, y_trainCAN, y_testCAN = train_test_split(x_canada, y_canada, test_size=0.3, random_state=0)
X_trainFRA, X_testFRA, y_trainFRA, y_testFRA = train_test_split(x_france, y_france, test_size=0.3, random_state=0)
X_trainJPN, X_testJPN, y_trainJPN, y_testJPN = train_test_split(x_japan, y_japan, test_size=0.3, random_state=0)
X_trainUK, X_testUK, y_trainUK, y_testUK = train_test_split(x_uk, y_uk, test_size=0.3, random_state=0)
X_trainUS, X_testUS, y_trainUS, y_testUS = train_test_split(x_us, y_us, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

rgr_world = LinearRegression()
rgr_can = LinearRegression()
rgr_fra = LinearRegression()
rgr_jpn = LinearRegression()
rgr_uk = LinearRegression()
rgr_us = LinearRegression()

# Train the model using the training sets
rgr_world.fit(X_train, y_train)
rgr_can.fit(X_trainCAN, y_trainCAN)
rgr_fra.fit(X_trainFRA, y_trainFRA)
rgr_jpn.fit(X_trainJPN, y_trainJPN)
rgr_uk.fit(X_trainUK, y_trainUK)
rgr_us.fit(X_trainUS, y_trainUS)

# Printing Coefficients, MSE, and Variance Score
print('Estimated World Avg Coefficients: \n', rgr_world.coef_)
print("Estimated World Avg Coefficients MSE: %.2f"
     % np.mean((rgr_world.predict(X_train) - y_train) ** 2))
print('Estimated World Avg Variance Score: %.2f\n' %rgr_world.score(X_train, y_train))

print('Canada Coefficients: \n', rgr_can.coef_)
print("Canada MSE: %.2f"
     % np.mean((rgr_can.predict(X_trainCAN) - y_trainCAN) ** 2))
print('Canada Variance Score: %.2f\n' %rgr_can.score(X_trainCAN, y_trainCAN))

print('France Coefficients: \n', rgr_fra.coef_)
print("France MSE: %.2f"
     % np.mean((rgr_fra.predict(X_trainFRA) - y_trainFRA) ** 2))
print('France Variance Score: %.2f\n' %rgr_fra.score(X_trainFRA, y_trainFRA))

print('Japan Coefficients: \n', rgr_jpn.coef_)
print("Japan MSE: %.2f"
     % np.mean((rgr_jpn.predict(X_trainJPN) - y_trainJPN) ** 2))
print('Japan Variance Score: %.2f\n' %rgr_jpn.score(X_trainJPN, y_trainJPN))

print('Canada Coefficients: \n', rgr_uk.coef_)
print("Canada MSE: %.2f"
     % np.mean((rgr_uk.predict(X_trainUK) - y_trainUK) ** 2))
print('Canada Variance Score: %.2f\n' %rgr_uk.score(X_trainUK, y_trainUK))

print('Canada Coefficients: \n', rgr_us.coef_)
print("Canada MSE: %.2f"
     % np.mean((rgr_us.predict(X_trainUS) - y_trainUS) ** 2))
print('Canada Variance Score: %.2f\n' %rgr_us.score(X_trainUS, y_trainUS))
```

```python
# 5 Countries + The World Linear Regression Graph
# Plot Outputs
plt.scatter(x_world, y_world, color="black")
plt.plot(X_test, rgr_world.predict(X_test), color='purple', linewidth=2, label = 'The World')
plt.plot(X_testCAN, rgr_can.predict(X_testCAN), color='blue', linewidth=2, label = 'Canada')
plt.plot(X_testFRA, rgr_fra.predict(X_testFRA), color='green', linewidth=2, label = 'France')
plt.plot(X_testJPN, rgr_jpn.predict(X_testJPN), color='yellow', linewidth=2, label = 'Japan')
plt.plot(X_testUK, rgr_uk.predict(X_testUK), color='orange', linewidth=2, label = 'United Kingdoms')
plt.plot(X_testUS, rgr_us.predict(X_testUS), color='red', linewidth=2, label = 'America')

# Label Graph
plt.xlabel('Money Spent on Health Care in USD')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy vs Money Spent on Health Care')
plt.legend(loc="lower right")
plt.savefig('LIN_REG')
```
